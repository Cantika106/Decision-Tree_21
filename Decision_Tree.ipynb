```python
import pandas as pd
df= pd.read_csv("/content/Dataset_Weather.csv")
df
```

```python
inputs = df.drop('tgl',axis='columns')
target = df['tgl']
inputs
```

```python

from sklearn.preprocessing import LabelEncoder
```

```python
le_Tahun = LabelEncoder()
le_Bulan = LabelEncoder()
le_temp7 = LabelEncoder()
```

```python
inputs['Tahun_n'] = le_Tahun.fit_transform(inputs['Tahun'])
inputs['Bulan_n'] = le_Bulan.fit_transform(inputs['Bulan'])
inputs['temp7_n'] = le_temp7.fit_transform(inputs['temp7'])
inputs.head()
```

```python
inputs_n = inputs.drop(['Tahun', 'Bulan', 'temp7'], axis='columns')
inputs_n
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
inputs_n['Tahun_n'] = inputs_n['Tahun_n'].astype(int)
inputs_n['Bulan_n'] = inputs_n['Bulan_n'].astype(int)
inputs_n['temp7_n'] = inputs_n['temp7_n'].astype(int)

model.fit(inputs_n, target)
```
